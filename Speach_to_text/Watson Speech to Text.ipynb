```python
files = []
for filename in os.listdir('.'):
    if filename.endswith(".mp3") and filename !='audio_08-31.mp3':
        files.append(filename)
files.sort()
```

## 3. Perform Conversion

```python
results = []
for filename in files:
    with open(filename, 'rb') as f:
        res = stt.recognize(audio=f, content_type='audio/mp3', model='es-CO_NarrowbandModel', continuous=True, \
                           inactivity_timeout=360).get_result()
        results.append(res)
```

```python
text = []
for file in results:
    for result in file['results']:
        text.append(result['alternatives'][0]['transcript'].rstrip() + '.\n')
```

```python
with open('output.txt', 'w') as out:
    out.writelines(text)
```
