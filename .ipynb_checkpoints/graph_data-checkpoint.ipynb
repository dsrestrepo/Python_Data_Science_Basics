# Descripción:

### Atributos (columnas 1-13)

1. age
2. sex
3. chest pain type (4 values)
4. resting blood pressure
5. serum cholestoral in mg/dl
6. fasting blood sugar > 120 mg/dl
7. resting electrocardiographic results (values 0,1,2)
8. maximum heart rate achieved
9. exercise induced angina
10. oldpeak = ST depression induced by exercise relative to rest
11. the slope of the peak exercise ST segment
12. number of major vessels (0-3) colored by flourosopy
13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

### Etiqueta (columna 14)

14. Labels: 1: enfermedad cardiaca; 0: saludable

# Matplotlib

# Gráfico de puntos

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Creamos figura 
plt.figure(figsize=(10, 6))

# Creamos diccionario de clase : color
colores = {1:'red', 0:'blue'}

# Graficamos diagrama de puntos
plt.scatter(x=disease.age, y = disease.thalach, c=disease.target.map(colores))
plt.xlabel('Edad')
plt.ylabel('Ritmo cardiaco máximo')

# Añadimos el parche con información de las gráficas
red_patch = mpatches.Patch(color='red', label='negative')
blue_patch = mpatches.Patch(color='blue', label='positive')

plt.legend(handles=[red_patch, blue_patch])

# guardar figura
plt.savefig('ejemplo.png')

# mostrar
plt.show()
```

```python
# Creamos instancia de la figura
fig = plt.figure(figsize=(10,7))

# definimos la proyeccion de la figura (3D)
ax = fig.add_subplot(111, projection='3d')

# como es 3D requiere 3 ejes y los colores se dan por las clases
ax.scatter(xs=disease.age, ys=disease.thalach, zs=disease.trestbps, c=disease.target.map(colores), s=60)

# definimos los ejes 'x', 'y' y 'z'
ax.set_xlabel('edad', fontsize=10, rotation=-20)
ax.set_ylabel('Ritmo Cardiaco máximo', fontsize=10, rotation=30)
ax.set_zlabel('presión arterial en reposo', fontsize=10, rotation=87)

# Rojos negativos 
# Azul positivos
red_patch = mpatches.Patch(color='red', label='negative')
blue_patch = mpatches.Patch(color='blue', label='positive')

plt.legend(handles=[red_patch, blue_patch])


# Mostramos
plt.show()
```

# Mapa de Calor

```python
def matriz_correlacion(df):
    # Calcular la correlación
    corr = df.corr(method='pearson')
    # Tamaño de la gráfica
    plt.figure(figsize=(15, 8))
    # Mostrar gráfica como mapa de calor con valores
    sns.heatmap(corr, annot=True)
    plt.show()
```

```python
# Calculamos la correlación de pearson:
matriz_correlacion(disease)
```
