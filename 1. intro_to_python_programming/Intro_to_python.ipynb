# loops

### For Loop

```python
names = ['Raul', 'David', 'Laura', 'Lucia', 'Daniel', 'Roberto', 'Leo']

# for loop in a list
for name in names:
  print(name)

print('')

# For loop in a range:
for i in range(len(names)):
  name = names[i]
  print(name)
```

### While loop

```python
i = int(input('say a value: '))
arr = []

while i > 0 and i < 10:
  item = input('say an item: ')
  arr.append(item)

  i-=1

print('The list with items is: ', arr)
```

```python
size = int(input('say a value: '))
arr = []

for i in range(size):
  item = input('say an item: ')
  arr.append(item)
  
print('The list with items is: ', arr)
```

# Example IMC

$$
BMI = \frac{weight [kg]} {(height [m])^ 2 }
$$

```python
class Person:
  def __init__(self, name, weight, height):
    self.name = name
    self.weight = weight
    self.height = height
    self.bmi = weight / (height ** 2)

  def get_bmi(self):
    self.bmi = self.weight / (self.height ** 2)
    print(f'The bmi of {self.name} is: {self.bmi}%')
```

```python
David = Person('David', 74, 1.80)
David.get_bmi()
```

```python
David.height = 1.70
print(David.bmi)
print(David.get_bmi())
```
